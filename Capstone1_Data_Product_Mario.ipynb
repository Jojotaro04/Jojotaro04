```python
import pandas as pd #Menggunakan fungsi pandas
import numpy as np #Menggunakan library numpy

data = '/content/drive/MyDrive/Datacsv/All_Appliances.csv' #path data csv dari google drive

data_appl = pd.read_csv(data)   #Membaca file CSV dengan encoding

data_appl #print data
```

DATA TRANSFORMATION

```python
data_appl['Number_of_Ratings'] = data_appl['no_of_ratings'].replace(np.NaN, 0) # mengganti NaN menjadi 0 di kolom no_of_ratings
data_appl[['no_of_ratings', 'Number_of_Ratings']]
```

```python
data_appl['Ratings'] = data_appl['ratings'].replace(np.NaN, 0) # mengganti NaN menjadi 0 di kolom ratings
data_appl[['ratings', 'Ratings']]
```

```python
data_appl.fillna(0) #Mengisi data kosong
```

```python
data_appl.drop_duplicates() #Menghapus data duplikat
```

```python
# Clean the 'discount_price' column
data_appl['discount_priceF'] = data_appl['discount_price'].replace({'・E・Eｹ': '', '₹': '', ',': ''}, regex=True).astype(float)

# Display the relevant columns
print(data_appl[['name', 'discount_priceF']].head())
```

```python

data_appl['actual_priceF'] = (
    data_appl['actual_price']
    .astype(str)  #merubah trelebih dahulu menjadi sting
    .str.replace(r'[^\d]', '', regex=True)  # #meghilangkan karakter non numeric
    .replace('', np.nan)  #menganti string dengan nan
   .astype(float)  # diubah kembali menjadi tipe data floar
)

print(data_appl[['name', 'actual_priceF']])
```

```python
data_appl.columns = data_appl.columns.str.upper() #Mengubah nama kolom menjadi huruf kapital
data_appl
```

```python
data_appl.columns = data_appl.columns.str.lower() #Mengubah nama kolom menjadi huruf biasa
data_appl
```
